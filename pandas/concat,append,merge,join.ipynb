```python
import pandas as pd
data1={'Name':['adhi','biju','ravi','unni'],
       'age':[20,21,22,23],
       'address':['kochi','kollam','kannur','kozhi'],
       'qualification':['msc','bsc','b.com','b.tech']}
data2={'Name':['adhi','kiran','anandhu','pindam'],
       'age':[20,30,33,24],
       'address':['kochi','chennai','goa','kozhi'],
       'qualification':['msc','b.com','msc','b.tech']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[1,2,3,4])
print(df,'\n\n',df1)

              
```

```python
#concatination
rslt=pd.concat([df,df1])
rslt
```

```python
#setting logic on axis
rslt1=pd.concat([df,df1],axis=1,join='inner')
rslt1
```

```python
#outer join
rslt1=pd.concat([df,df1],axis=1,sort=False)
rslt1
```

```python
#append
rslt=df.append(df1)
rslt
```

```python
#merging

import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'Name':['adhi','biju','ravi','unni'],
       'age':[20,21,22,23]}
data2={'key':['k1','k2','k3','k4'],
       'address':['kochi','chennai','goa','kozhi'],
       'qualification':['msc','b.com','msc','b.tech']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[1,2,3,4])
print(df,'\n\n',df1)
```

```python
#merging
rslt=pd.merge(df,df1,on="key")
rslt
```

```python
#joinig with multiple keys
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'key1':['k0','k1','k1','k0'],
       'Name':['adhi','biju','ravi','unni'],
       'age':[20,21,22,23]}
data2={'key':['k1','k2','k3','k4'],
       'key1':['k0','k1','k0','k1'],
       'address':['kochi','chennai','goa','kozhi'],
       'qualification':['msc','b.com','msc','b.tech']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,'\n\n',df1)
```

```python
rslt=pd.merge(df,df1,on=['key','key1'])
rslt
```

```python
#how=left
rslt=pd.merge(df,df1,how='left',on=['key','key1'])
rslt
```

```python
#how=right
rslt=pd.merge(df,df1,how='right',on=['key','key1'])
rslt
```

```python
#how=outer
rslt=pd.merge(df,df1,how='outer',on=['key','key1'])
rslt
```

```python
#how=inner
rslt=pd.merge(df,df1,how='inner',on=['key','key1'])
rslt
```

```python
#single index
import pandas as pd
data1={
 
       'Name':['adhi','biju','ravi','unni'],
       'age':[20,21,22,23]}
data2={

       'address':['kochi','chennai','goa','kozhi'],
       'qualification':['msc','b.com','msc','b.tech']}
df=pd.DataFrame(data1,index=pd.Index(['k0','k1','k2','k3']))
index = pd.MultiIndex.from_tuples([('K0', 'Y0'), ('K1', 'Y1'),
                                   ('K2', 'Y2'), ('K2', 'Y3')],
                                   names=['key', 'Y'])
df1=pd.DataFrame(data2,index=index)

print(df,'\n\n',df1)

rslt1=df.join(df1,how='inner')
rslt1
```
