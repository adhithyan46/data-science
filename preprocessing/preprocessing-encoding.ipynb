```python
import pandas as pd
#one-hot encoding
df=pd.DataFrame({'colors':['red','green','blue','orange','yellow']})
print('Dataframe before encoding')
print(df['colors'])
one_hot=pd.get_dummies(df['colors'])
# df1=pd.concat([df,one_hot],axis=1)
# df1=df1.drop('colors',axis=1)
print('Dataframe after encoding')
one_hot
```

```python
#dummy-encoding
print('Before encoding')
print(df['colors'])
dummy=pd.get_dummies(df['colors'],drop_first=True)
print('After encoding')
dummy
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
print('Befor encoding')
print(df['colors'])
df['color_label']=le.fit_transform(df['colors'])
print('After encoding')
df
```

```python
# one hot encoding
df=pd.read_csv('cardata.csv')
print(f'before encoding \n{df.head(20)}')
ctgry_cols=['Car_Name','Fuel_Type','Seller_Type','Transmission']
one_hot=pd.get_dummies(df ,'ctgry_cols')
print('\nafter encoding')
one_hot.head()
```

```python
#label encoding
df=pd.read_csv('cardata.csv')
ctgry_cols=['Fuel_Type','Seller_Type','Transmission']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for cols in ctgry_cols:
    df[cols]=le.fit_transform(df[cols])
print(df.head())
```
