```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.distplot([0,1,2,3,4,5],hist=False)
plt.show()
```

```python
#normal distribution
x=random.normal(loc=1,scale=2,size=(2,3))
x
```

```python
sns.distplot(random.normal(size=1000),hist=False)
plt.show()
```

```python
#binomial distribution(discreat)
x=random.binomial(n=10,p=0.5,size=(2,3))
x
```

```python
sns.distplot(random.binomial(n=10,p=0.5,size=1000),hist=True,kde=False)
```

```python
#difference between normal and binomial
sns.distplot(random.normal(loc=50,scale=5,size=1000),hist=False,label='normal')
sns.distplot(random.binomial(n=100,p=0.5,size=1000),hist=False,label="binomial")
plt.legend()
```

```python
#poisson distribution
x=random.poisson(lam=2,size=(2,3))
x
```

```python
sns.distplot(random.poisson(lam=2,size=1000),kde=False)
```

```python
#difference between poisson and normal
sns.distplot(random.normal(loc=50,scale=7,size=1000),hist=False,label="normal")
sns.distplot(random.poisson(lam=20,size=1000),hist=False,label="poisson")
plt.legend()
```

```python
# difference between binomial and poisson
sns.distplot(random.binomial(n=1000,p=0.05,size=1100),hist=False,label='binomial')
sns.distplot(random.poisson(lam=20,size=1000),hist=False,label='poisson')
plt.legend()
```

```python
#uniform
sns.distplot(random.uniform(size=1000),hist=False)
```
