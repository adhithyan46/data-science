```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('TV_show_data.csv')
df.head()
```

```python
##Histogram
# x=df['Rating']
y=df['Language']
# plt.hist(x)
plt.hist(y)
```

```python
#pie chart
y=np.array([30,25,15,20,10])
plt.figure(figsize=(9,9))
mylabels=['apple','orange','kiwi','grapes','banana']
myexplode=[0,0,0,0,0.1]#defining explode value
mycolor=['red','green','blue','cyan','orange']#defining color
plt.pie(y,labels=mylabels,startangle=90,autopct='%1.1f%%',explode=myexplode,shadow=True,colors=mycolor)
plt.legend(title='fruits name')#legend with title
plt.show()
```
