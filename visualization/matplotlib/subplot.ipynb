```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,2,4,5])
y=np.array([1,2,3,5])

plt.subplot(1,2,1)
plt.plot(x,y,c='r')
plt.grid(ls='--')
plt.title('rating',loc="left")

x1=np.array([0,3,2,4,])
y1=np.array([0,1,3,4])

plt.subplot(1,2,2)

plt.plot(x1,y1,lw='5',c='g')
plt.grid(ls='--')

plt.title('show',loc="left")

plt.suptitle('tv show')
plt.savefig('tv show.png',dpi=300)
plt.show()
```
