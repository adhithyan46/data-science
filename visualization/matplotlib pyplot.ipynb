```python
import matplotlib.pyplot as plt
apple=[0.21,0.99,1.33]
plt.plot(apple)
plt.show()
```

```python
#axis value
year=[2019,2020,2013]
apple=[0.21,0.99,1.33]
plt.plot(year,apple)
plt.show()
```

```python
#setting label,title,legend
year=[2019,2020,2013]
apple=[0.21,0.99,1.33]
plt.plot(year,apple)
plt.xlabel('year')
plt.ylabel('apple')
plt.title('fruit graph')
plt.legend(['year','apple'])
plt.show()
```

```python
#multiple graph
year=[2019,2020,2013]
apple=[0.21,0.99,1.33]
orange=[0.33,0.55,0.98]
plt.plot(year,apple,marker='o')
plt.plot(year,orange,marker='*')
plt.xlabel('year')
plt.ylabel('fruit')
plt.legend(['apple','orange'])
plt.show()
```

```python
#markers
year=[2019,2020,2013]
apple=[0.21,0.99,1.33]
orange=[0.33,0.55,0.98]
plt.plot(year,apple,marker='o')
plt.plot(year,orange,marker='x')
plt.xlabel('year')
plt.ylabel('fruit')
plt.show()
```
