```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=sns.load_dataset('tips')
df.head(3)
```

```python
numeric_cols=['total_bill','tip','size'] #selecting numerical columns
corr=df[numeric_cols].corr()#correlation of the columns
sns.heatmap(corr,annot=True)
```

```python
df=pd.read_csv('TV_show_data.csv')
df.info()
```

```python
num_cols=['Total Seasons','Total Episodes','Average Runtime','Rating']
corr=df[num_cols].corr()
sns.heatmap(corr,annot=True)
```

```python
sns.heatmap(corr,annot=True,fmt='1g')
```

```python
#adjusting axis
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0)
```

```python
#colourmap
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral')
```

```python
#linewidth,linecolor
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',linewidth='3',linecolor='white')
```

```python
# orientation
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',cbar_kws={'orientation':'horizontal'})
```

```python
#shape
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',square=True)
```

```python
# changing matrix shape
matrix=np.triu(corr) #triu (lower triangle)
matrix1=np.tril(corr) #tril (upper triangle)
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',mask=matrix)
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0,cmap='Spectral',mask=matrix1)
```
