```python
import pandas as pd
df=pd.read_csv('cardata.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

col=['Seller_Type','Transmission','Fuel_Type','Car_Name']
encoder = LabelEncoder()

# Apply LabelEncoder to specified columns
for col in col:
    df[col] = encoder.fit_transform(df[col])
df
```

```python
input=df.drop('Selling_Price',axis=1)
target=df['Selling_Price']
input.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(input,target,test_size=0.3,random_state=100)
print('x_trait\n',X_train.head())
print('x_test\n',X_test.head())
print('y_trait\n',y_train.head())
print('y_test\n',y_test.head())
```

```python
from sklearn.preprocessing import StandardScaler

cols=['Year','Kms_Driven','Owner']
scaler = StandardScaler()

# Apply StandardScaler to numerical columns
df[cols] = scaler.fit_transform(df[cols])
df
```

```python
from sklearn import tree
model=tree.DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_train)
```
